```python
from datasets import load_dataset

tiny_stories = load_dataset("roneneldan/TinyStories")
```

```python
limit = 100000
```

```python
import os
train_file_path = os.path.join("data", "tiny_stories_train.txt")
print(f"Writing training data to {train_file_path}...")
with open(train_file_path, "w", encoding="utf-8") as f:
    for i, item in enumerate(tiny_stories["train"]):
        if i >= limit:
            break
        f.write(item["text"] + "\n\n")
val_file_path = os.path.join("data", "tiny_stories_val.txt")
print(f"Writing validation data to {val_file_path}...")
with open(val_file_path, "w", encoding="utf-8") as f:
    limit = 5000
    for i, item in enumerate(tiny_stories["validation"]):
        if i >= limit:
            break
        f.write(item["text"] + "\n\n")
```
